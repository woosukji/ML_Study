# Softmax Classification

## 6-1 Zoo Data

```python
import tensorflow as tf
import tensorflow.keras as k
import numpy as np
```

```python
data = np.loadtxt('zoo.csv', delimiter = ',')
x = data[:, 0:-1]
y = data[:, [-1]]
y_one_hot = tf.reshape(tf.one_hot(y,7), [-1,7])
```

```python
model = k.Sequential()
model.add(k.layers.Dense(7, input_dim = 16))
model.add(k.layers.Activation('softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer = k.optimizers.SGD(lr = 0.1), metrics = ['acc'])
model.summary()
```

```python
model.fit(x[10:], y_one_hot[10:], epochs = 1000)
```

```python
res = model.predict_classes(data[:10, :-1]) == data[:10, -1]
print('Accuracy : {}%'.format(np.sum(res)*10))
```

## 7-2 MNIST Data

```python
mnist = k.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()y_train_one_hot = tf.one_hot(y_train, 10)
```

```python
x_train_flat = tf.reshape(x_train, [-1,784])
y_train_one_hot = tf.one_hot(y_train, 10)
```

```python
mini_model = k.Sequential()
mini_model.add(k.layers.Dense(units = 10, input_dim = 784, activation = 'softmax'))
mini_model.compile(loss = 'categorical_crossentropy', optimizer = k.optimizers.SGD(lr = 0.1, metrics = ['acc'])
mini_model.summary()
```

```python
mini_model.fit(x_train_flat, y_train_one_hot, epochs = 15)
```

```python
x_test_flat = tf.reshape(x_test, [-1,784])

np.mean(mini_model.predict_classes(x_test_flat) == y_test) * 100
```
