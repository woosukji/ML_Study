# Convolutional Neural Network

```python
import tensorflow as tf
import tensorflow.keras as k
import numpy as np
```

```python
(x, y), (x_t, y_t) = k.datasets.mnist.load_data()
x, x_t = x / 255, x_t / 255
x, x_t = x.reshape(-1, 28, 28, 1), x_t.reshape(-1, 28, 28, 1)
y, y_t = tf.one_hot(y, 10), tf.one_hot(y_t, 10)
```

```python
model = k.Sequential()
model.add(k.layers.Conv2D(32, (3,3), (1,1), 'same', input_shape = (28,28,1), activation = 'relu'))
model.add(k.layers.MaxPool2D((2,2), None, 'same'))

model.add(k.layers.Conv2D(64, (3,3), (1,1), 'same', activation = 'relu'))
model.add(k.layers.MaxPool2D((2,2), None, 'same'))

model.add(k.layers.Conv2D(128, (3,3), (1,1), 'same', activation = 'relu'))
model.add(k.layers.MaxPool2D((2,2), None, 'same'))

model.add(k.layers.Flatten())
model.add(k.layers.Dense(625, 'relu', kernel_initializer = 'glorot_normal'))
model.add(k.layers.Dropout(0.3))
model.add(k.layers.Dense(10, 'softmax', kernel_initializer = 'glorot_normal'))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['acc'])
model.summary()
```

```python
model.fit(x, y, batch_size = 128, epochs = 12) 
```

```python
model.evaluate(x_t, y_t)
```

```python
import matplotlib.pyplot as plt
```

```python
y_s = np.argmax(y_t, axis = -1)
wrong_index = np.where(model.predict_classes(x_t) != y_s)
```

```python
images = x_t[wrong_index].reshape(-1,28,28)
prediction = model.predict(x_t[wrong_index])
# truth = np.argmax(y_t[wrong_index], axis = -1)  -  doesn't work : can't slice tf tensor wish tuples
truth = y_s[wrong_index]
```

```python
plt.figure(figsize = (2*6, 2*5))
for i in range(15) :
    predict_num = np.argmax(prediction[i])
    
    plt.subplot(5, 6, 2*i+1)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(images[i], cmap = plt.cm.binary)
    plt.xlabel('{} {:2.0f}% ({})'.format(
        predict_num,
        np.max(prediction[i]) * 100,
        truth[i]
    ), color = 'blue' if predict_num == truth[i] else 'red')
    
    plt.subplot(5, 6, 2*i+2)
    plt.grid(False)
    plt.xticks(range(10))
    plt.yticks([])
    thisplt = plt.bar(range(10), prediction[i], color="#777777")
    plt.ylim(0,1)
    thisplt[predict_num].set_color('red')
    thisplt[truth[i]].set_color('blue')

plt.suptitle('Sample of 98 Wrong Predictions out of 10,000 test cases')
```
